```python
%matplotlib inline
import pandas as pd
```

```python
df = pd.read_excel('data/policyagendas/media1960-2008_websiteversion_111007.xlsx',sheet_name='Sheet1')
df.head()
```

```python
df.groupby('Date')['Title'].count().plot()
```

```python
df['Year'] = df['Date'].apply(lambda x: x.year)
```

```python
df.groupby('Year')['Title'].count().plot()
```

```python
topic_counts = df.groupby(['Year','Major_Topic'])['Title'].count().unstack()
```

```python
topic_counts[12].plot()
```

```python
totals = topic_counts.sum(axis=1)
```

```python
(topic_counts[12] / totals).plot()
(topic_counts[2] / totals).plot()
(topic_counts[16] / totals).plot()
```
